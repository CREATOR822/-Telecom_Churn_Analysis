# prepare data for training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 11)
```

# applying multiple models at same time

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LogisticRegression
```

```python
model = [DecisionTreeRegressor,SVR,RandomForestRegressor,KNeighborsRegressor,AdaBoostRegressor,LogisticRegression]

for mod in model:
    reg = mod()
    reg = reg.fit(x_train,y_train)
    print(mod , 'accuracy',reg.score(x_test,y_test))
```

# applying K-fold 

```python
from sklearn.model_selection import KFold,cross_val_score
```

# Here is loop for multiple model accuracy

```python
models = []
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('RF', RandomForestRegressor()))
models.append(('SVM', SVR()))
models.append(('AdaBoost', AdaBoostRegressor()))
models.append(('lgr', LogisticRegression()))


results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10)
    cv_result =cross_val_score(model,x_train,y_train,cv=kfold)
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean())
```

# test score on logistic reg

```python
lgr=LogisticRegression(max_iter=6000, random_state=67)
lgr.fit(x_train, y_train)
lgr.score(x_test, y_test)
```

```python
y_pred = lgr.predict(x_test)
```

```python
lgr.score(x_test, y_pred)
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

```python
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,classification_report, confusion_matrix
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
